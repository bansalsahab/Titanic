```python
# q1 For the given dataset, calculate the variance, standard deviation, and interquartile range (IQR) of the
# salaries of employees in the HR and Marketing departments.

#claculating variance
hr_department = dataset[dataset['Department'] == 'hr']
hr_salary = hr_department['Salary_INR']
mar_department=dataset[dataset['Department']=='marketing']
mar_salary=mar_department['Salary_INR']
```

```python
mar_var=np.var(mar_salary)
```

```python
hr_var=np.var(hr_salary)
#round the variance accordingly
```

```python
#calculating standard devation
hr_sal_sd=np.std(hr_salary)
mar_sal_sd=np.std(mar_salary)
```

```python
# interquaratile range
hr_sal_iqr = np.percentile(hr_salary, 75) - np.percentile(hr_salary, 25)
mar_sal_iqr=np.percentile(mar_salary,75)-np.percentile(mar_salary,25)
```
